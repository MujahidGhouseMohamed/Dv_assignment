```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("IPL_BallByBall2008_2024.csv")
data.dropna(inplace=True)
data.reset_index(drop=True, inplace=True)
data
```

```python
df = pd.DataFrame(data)

# Visualization 1: Bar plot showing the total runs scored by each batting team
plt.figure(figsize=(10, 6))
sns.barplot(x='Batting team', y='score', data=df, estimator=sum)
plt.title('Total Runs Scored by Each Batting Team')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 2: Line plot showing the trend of runs scored over different innings
plt.figure(figsize=(10, 6))
sns.lineplot(x='Innings No', y='score', data=df, marker='o')
plt.title('Trend of Runs Scored Over Different Innings')
plt.xlabel('Innings')
plt.ylabel('Runs Scored')
plt.show()
```

```python
# Visualization 3: Pie chart showing the distribution of wicket types
plt.figure(figsize=(8, 8))
df['wicket_type'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Distribution of Wicket Types')
plt.ylabel('')
plt.show()
```

```python
# Visualization 4: Box plot showing the distribution of runs scored by each batting team
plt.figure(figsize=(10, 6))
sns.boxplot(x='Batting team', y='score', data=df)
plt.title('Distribution of Runs Scored by Each Batting Team')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 5: Count plot showing the frequency of each type of extras
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='type of extras')
plt.title('Frequency of Each Type of Extras')
plt.xlabel('Type of Extras')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 6: Scatter plot showing the relationship between runs scored and innings number
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Innings No', y='score', data=df, hue='Batting team', style='Batting team')
plt.title('Relationship between Runs Scored and Innings Number')
plt.xlabel('Innings')
plt.ylabel('Runs Scored')
plt.show()
```

```python
# Visualization 7: Heatmap showing the correlation between different numerical features
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Different Numerical Features')
plt.show()
```
